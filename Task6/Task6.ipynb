```python
#TASK 1#
def solution (t):
    N = int(input())
    b = ''
    while N > 0:
        b = str(N % 2) + b
        N = N // 2
    c = list(map(int, str(b)))
    m = len(list(c))-1
    t = 0
    f = 0
    for i in range (0,m):
        while m > 1:
            if c[m] - c[m-1] >= 0:
                f += c[m] - c[m-1]
                t += 1
                if f == 0:
                    t = 0
            else:
                t = 0
            m = m-1
    #print(t)

    
    return t
```
